```python
import pandas as pd
df = pd.read_csv("intraday_5min_IBM.csv", index_col='timestamp', parse_dates=True)
df
```

```python
df.index
```

```python
# Timezone-aware time index
df=df.tz_localize(tz="US/Eastern")
df.index
```

```python
# convert to Moscow time
df = df.tz_convert(tz="Europe/Moscow")
df
```

```python
# Getting all timezones
from pytz import all_timezones
all_timezones
```

```python
# tz_convert is applicable to df.index as well
df.index.tz_convert(tz="Europe/Helsinki")
```

```python
# Create date range
rng = pd.date_range(start='1/1/2020', periods=10, freq='H') # not timezone aware
rng
```

```python
# using pytz timezone 
rng = pd.date_range(start='1/1/2020', periods=10, freq='H', tz="Europe/London")
rng
```

```python
# using dateutil - compare result's dtype with previous
rng = pd.date_range(start='1/1/2020', periods=10, freq='H', tz="dateutil/Europe/London")
rng
```

```python
# Time series  enumerated
rng = pd.date_range(start='1/1/2020 09:30:00', periods=48, freq='30min')
s = pd.Series(range(48), index=rng)
s
```

```python
# convert to Mosccow
msc = s.tz_localize(tz='Europe/Moscow')
msc.index
```

```python
# New York
ny = s.tz_localize(tz='America/New_York')
ny.index
```

```python
# if you add those series, it will add only matching times
print((ny + msc)[0:48])

```
