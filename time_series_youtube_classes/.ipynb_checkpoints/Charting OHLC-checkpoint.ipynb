```python
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf

df = pd.read_csv("intraday_5min_IBM.csv", index_col='timestamp', parse_dates=True)
```

```python
# Re-sort in ascending order
df = df.sort_index(ascending=True)
df
```

```python
df.index.name = 'Time'
df.shape
df
```

```python
# mplfinance expects specific column names (?)
# open ==> Open, etc
df = df.rename(columns={"open":"Open"
                        , "high":"High"
                        , "low":"Low"
                        , "close":"Close"
                        , "volume":"Volume"
                       })
mpf.plot(df)
```

```python
# Candlestick:
mpf.plot(df["2020-05-21"], type='candle')
```

```python
# All Types
for tp in ('candle', 'line', 'renko', 'pnf', 'bars'):
    print(tp)
    mpf.plot(df["2020-05-22"], type=tp)
```

```python
# Adding Moving Average and Volume
mpf.plot(df["2020-05-21"], type='candle', mav=(3,6,9), volume=True)
```

```python
# Customization: title and labels
mpf.plot(df["2020-05-21"]
         , type='candle'
         , volume=True
         , title='\nIBM 2020-05-21'
         , ylabel='OHLC Candles'
         , ylabel_lower='Shares\nTraded')

```

```python
# Customization: Styles
mpf.available_styles()
```

```python
date = "2020-05-22"
kwargs = dict(type='candle'
              , mav=(10, 15)
              , volume=True
              , figratio=(16,8)
              , figscale=0.75)
eod_df = df[date]
for stl in mpf.available_styles():
    kwargs['title'] = "\nIBM {} \"{}\" style".format(date, stl)
    mpf.plot(eod_df, **kwargs, style=stl)
```

```python
date = "2020-05-22"
kwargs = dict(type='candle'
              , mav=(5, 10, 15)
              , volume=True
              , figratio=(16,8)
              , figscale=0.75)
eod_df = df[date]
mpf.plot(eod_df, **kwargs, style='sas', title="\nIBM {}".format(date))
```

```python
# addplot
date = "2020-05-22"
kwargs = dict(type='candle'
#               , mav=(5, 10, 15) # no moving average to plot
              , volume=True
              , figratio=(16,8)
              , figscale=0.75
              , style='sas'
              , title="\nIBM {}".format(date))
eod_df = df[date]

# average between Open and Close
mpf.plot(eod_df
         , **kwargs
         , addplot=mpf.make_addplot(pd.DataFrame((eod_df.Open + eod_df.Close)/2, eod_df.index)))
```

```python
# add plot functionality

date = "2020-05-22"
kwargs = dict(type='candle'
              , volume=True
              , figratio=(16,8)
              , figscale=0.75
              , style='sas'
              , title="\nIBM {}".format(date))
eod_df = df[date]

addplts = [mpf.make_addplot(
             pd.DataFrame(eod_df.Close.rolling(10).mean(), eod_df.index))
         , mpf.make_addplot(
             pd.DataFrame(eod_df.Close.rolling(10).std(), eod_df.index))]

mpf.plot(eod_df, **kwargs, addplot=addplts) 
```

```python
# add plot functionality

date = "2020-05-22"
kwargs = dict(type='candle'
              , volume=True
              , figratio=(16,8)
              , figscale=0.75
              , style='default'
              , title="\nIBM {}".format(date))
eod_df = df[date]

stat_df = pd.DataFrame(index=eod_df.index)
stat_df["Mean"] = eod_df.Close.rolling(10).mean()
stat_df["Std"] = eod_df.Close.rolling(10).std()

mpf.plot(eod_df, **kwargs, addplot=[mpf.make_addplot(stat_df["Mean"], color='r')
                                    , mpf.make_addplot(stat_df["Std"], color='g')
                                   ]
        ) 

# more examples on https://github.com/matplotlib/mplfinance/blob/master/examples/addplot.ipynb

```

```python
# add plot functionality

date = "2020-05-22"
kwargs = dict(type='candle'
              , volume=True
              , figratio=(16,8)
              , figscale=0.75
              , style='default'
             )
eod_df = df[date]

stat_df = pd.DataFrame(index=eod_df.index)

for wnd in [5,10,15,20]:
    stat_df["Mean"] = eod_df.Close.rolling(wnd).mean()
    stat_df["Std"] = eod_df.Close.rolling(wnd).std()

    mpf.plot(eod_df, **kwargs, title="\nIBM {}. Stat. Window Size ={}".format(date, wnd)
             , addplot=[mpf.make_addplot(stat_df["Mean"], color='r')
                                    , mpf.make_addplot(stat_df["Std"], color='g')
                                   ]
        ) 

# more examples on https://github.com/matplotlib/mplfinance/blob/master/examples/addplot.ipynb

```
