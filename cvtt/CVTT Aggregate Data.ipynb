```python
# install mplfinance module
# pip install --upgrade mplfinance

# Examples, etc.
# https://github.com/matplotlib/mplfinance/tree/master/examples

# Add Plot examples
# https://github.com/matplotlib/mplfinance/blob/master/examples/addplot.ipynb
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf

fileName = "/localdisk/cvtt/aaa.csv"
col_names = pd.read_csv(fileName, nrows=0).columns
types_dict = {'Timestamp': str
              , 'AggrOpen': float
              , 'AggrHigh': float
              , 'AggrLow': float
              , 'AggrClose': float
              , 'AggrVolume': float
}
types_dict.update({col: str for col in col_names if col not in types_dict})
df = pd.read_csv(fileName, dtype=types_dict)

df["DateTs"] = df["Date"] + " " + df["Timestamp"]
df.index = pd.to_datetime(df["DateTs"], format='%Y%m%d %H%M%S.%f')
df.index.name = "tstamp"
df
```

```python
df = df.rename(columns={"AggrOpen":"Open"
                        , "AggrHigh":"High"
                        , "AggrLow":"Low"
                        , "AggrClose":"Close"
                        , "AggrVolume":"Volume"
                       })
kwargs = dict(type='candle'
              , mav=(10, 15)
              , volume=True
              , figratio=(16,8)
              , figscale=1
              , style='sas'
              , title=f"{df['Date'][0]} BTS/USD 5 min")
# for scale in [0.5, 0.75, 1, 2]:
#      kwargs['figscale'] = scale
#      mpf.plot(df, **kwargs)
mpf.plot(df, **kwargs)

```
